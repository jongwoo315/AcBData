```python
def PostData(url, _id, secret):
    post_args = '/posts/?key=value&access_token='+_id+'|'+secret
    post_url = url + post_args
    return post_url


def CommentData(url, post_id, _id, secret):
    return url + post_id + '/comments?limit=50&value=1&base_amount=1&access_token='+_id+'|'+secret


def render_to_json(url):
    web_response = urllib.request.urlopen(url)
    readable_page = web_response.read().decode('utf-8')
    json_data = json.loads(readable_page)
    return json_data
```
